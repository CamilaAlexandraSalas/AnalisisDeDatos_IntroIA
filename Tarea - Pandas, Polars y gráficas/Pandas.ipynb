## **Leyendo los archivos**

```python
import pandas as pd

green = pd.read_parquet('green_tripdata_2025-01.parquet')
yellow = pd.read_parquet('yellow_tripdata_2025-01.parquet')
```

```python
yellow
```

## **Promedio de ingresos brutos de cada tipo de taxi y de todos los taxis**

```python
TotalYellow = yellow['total_amount'].mean()
TotalGreen  = green['total_amount'].mean()

print("Promedio de ingresos brutos de cada tipo de taxi")
print("Promedio taxi Yellow: ", TotalYellow)
print("Promedio taxi Green: ", TotalGreen)
print("Promedio de todos los taxis: ", (TotalYellow + TotalGreen)/2)
```

## **Qué tipo de taxis recorrieron la mayor cantidad de distancia en Enero**

```python
DistanciaYellow = yellow['trip_distance'].sum()
DistanciaGreen  = green['trip_distance'].sum()

print("Distancia recorrida de cada tipo de taxi")
print("Distancia taxi Yellow: ", DistanciaYellow)
print("Distancia taxi Green: ", DistanciaGreen)

if DistanciaYellow > DistanciaGreen:
  print("El taxi Yellow recorrio mas distancia")
else:
  print("El taxi Green recorrio mas distancia")
```

# **Qué tipo de taxi recibe más propinas en promedio**

```python
PropinaYellow = yellow['tip_amount'].mean()
PropinaGreen  = green['tip_amount'].mean()

print("Propinas de cada tipo de taxi")
print("Propinas taxi Yellow: ", PropinaYellow)
print("Propinas taxi Green: ", PropinaGreen)

if PropinaYellow > PropinaGreen:
  print("El taxi Yellow recibe mas propinas")
else:
  print("El taxi Green recibe mas propinas")
```

## **Gráfico de barras reportando el importe del peaje total de cada tipo de Taxi**

```python
import matplotlib.pyplot as plt
```

```python
PeajeYellow = yellow['tolls_amount'].sum()
PeajeGreen  = green['tolls_amount'].sum()

print("Peajes de cada tipo de taxi")
print("Peaje taxi Yellow: ", PeajeYellow)
print("Peaje taxi Green: ", PeajeGreen)

tipos_taxi = ['Yellow Taxi', 'Green Taxi']
promedios_peaje = [PeajeYellow, PeajeGreen]

plt.bar(tipos_taxi, promedios_peaje, color=['yellow', 'green'])

plt.ylabel('Peaje')
plt.title('Importe de peaje por tipo de taxi')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
